## Columns and their description

1. **iso_code**-ISO 3166-1 alpha-3 three-letter country codes
2. **country**-Geographic location
3. **year**-Year of observation
4. **population**-Total population
5. **coal_electricity**-Electricity generation from coal measured in terawatt-hours
6. **gas_electricity-Electricity**- generation from gas, measured in terawatt-hours
7. **oil_electricity-Electricity**- generation from oil, measured in terawatt-hours
8. **nuclear_electricity**-Electricity generation from nuclear power, measured in terawatt-hours
9. **hydro_electricity**-Electricity generation from hydropower, measured in terawatt-hours
10.**solar_electricity**-Electricity generation from solar, measured in terawatt-hours
11.**wind_electricity**-Electricity generation from wind, measured in terawatt-hours
12.**biofuel_electricity**-Electricity generation from biofuels, measured in terawatt-hours
13.**other_renewable_exc_biofuel_electricity**-Electricity generation from other renewable sources excluding biofuels, measured      in terawatt-hours
14.**electricity_generation**-Electricity generation, measured in terawatt-hours
15.**electricity_demand**-Electricity demand, measured in terawatt-hours

## Feature Enggineering

```python
df['non_renewable_electricity']=df['coal_electricity']+df['gas_electricity']+df['oil_electricity']+df['nuclear_electricity']
```

```python
df['renewable_electricity']=df['hydro_electricity']+df['solar_electricity']+df['biofuel_electricity']+df['biofuel_electricity']+df['wind_electricity']+df['other_renewable_exc_biofuel_electricity']
```

```python
df['fossile_electricity']=df['coal_electricity']+df['gas_electricity']+df['oil_electricity']
```

```python
df['Non_fossile_electricity']=df['nuclear_electricity']+df['hydro_electricity']+df['solar_electricity']+df['biofuel_electricity']+df['wind_electricity']+df['other_renewable_exc_biofuel_electricity']
```
